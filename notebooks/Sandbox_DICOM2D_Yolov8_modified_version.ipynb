## Training

Training from yolov8n weights

Note: It could be possible change the checkpoint by our latest experiment. Just put the path of you model instead of yolov8n.pt in weights parameter Note: The dataset contains the config file (data.yaml). This config file associates the dataset previously downloaded. However, you can change it, when other datasets would be used.

```python
!yolo task=detect mode=train model=yolov8n.pt imgsz=640 data=/content/datasets/dicom_yolov8/data.yaml epochs=30 batch=8 name=yolov8n_v8_50e
```

## Validation

Run validation over the best model of our experiments.

```python
!yolo val model=/content/ultralytics/runs/detect/yolov8n_v8_50e/weights/best.pt data=/content/datasets/dicom_yolov8/data.yaml imgsz=640 batch=16
```
